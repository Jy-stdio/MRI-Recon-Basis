```python
import torch
from torch.linalg import pinv
from scipy.io import loadmat
import matplotlib.pyplot as plt
from IPython import display
%matplotlib inline
```

## FFT and IFFT

```python
def fft(data: torch.Tensor, norm: str = 'backward') -> torch.Tensor:
    data = torch.fft.ifftshift(data, dim=(-2, -1))
    data = torch.fft.fft2(data, norm=norm)
    data = torch.fft.fftshift(data, dim=(-2, -1))
    return data


def ifft(data: torch.Tensor, norm: str = 'backward') -> torch.Tensor:
    data = torch.fft.ifftshift(data, dim=(-2, -1))
    data = torch.fft.ifft2(data, norm=norm)
    data = torch.fft.fftshift(data, dim=(-2, -1))
    return data
```

## Mask

```python
# load and present sampling mask
cart_mask_path = "./data/smapling_mask/mask_cart_050.mat"
radial_mask_path = "./data/smapling_mask/mask_radial_010.mat"
random_mask_path = "./data/smapling_mask/mask_random_010.mat"

cart_mask = torch.tensor(loadmat(cart_mask_path)['mask_cart_050'])
radial_mask = torch.tensor(loadmat(radial_mask_path)['mask_radial_010'])
random_mask = torch.tensor(loadmat(random_mask_path)['mask_random_010'])

fig, axes = plt.subplots(1, 3, layout='compressed', figsize=(9, 3))

axes[0].set_title('Cartesian mask')
axes[0].imshow(cart_mask, cmap='gray')

axes[1].set_title('Radial mask')
axes[1].imshow(radial_mask, cmap='gray')

axes[2].set_title('Random mask')
axes[2].imshow(random_mask, cmap='gray')
```

## kspace visual and undersample

```python
def normalize(img, diff, range, eps=0.0):
    return (img - diff) / (range + eps)

def kspace_visualize(fft: torch.Tensor, kscale: int = 6):
    kspace = torch.abs(fft)
    scaled_kspace = torch.log1p(pow(10, kscale) * kspace)
    normalized_kspace = normalize(scaled_kspace, scaled_kspace.min(), scaled_kspace.max() - scaled_kspace.min())
    return normalized_kspace
```

## Image and undersampled image

```python
# load image, do undersampling and present
img_path = "./data/image_sos.mat"
img = loadmat(img_path)['DATA']
img = torch.tensor(img)

kspace = fft(img)
normalized_kspace = kspace_visualize(kspace)

cart_us_kspace = cart_mask * normalized_kspace
cart_us_img = torch.abs(ifft(cart_mask * fft(img)))

radial_us_kspace = radial_mask * normalized_kspace
radial_us_img = torch.abs(ifft(radial_mask * fft(img)))

random_us_kspace = random_mask * normalized_kspace
random_us_img = torch.abs(ifft(random_mask * fft(img)))

fig, axes = plt.subplots(2, 4, layout="compressed", figsize=(12, 6))

axes[0, 0].set_title('Original kspace')
axes[0, 0].imshow(normalized_kspace, cmap='gray')

axes[1, 0].set_title('Original image')
axes[1, 0].imshow(img, cmap='gray')

axes[0, 1].set_title('Cart Undersampled kspace')
axes[0, 1].imshow(cart_us_kspace, cmap='gray')

axes[1, 1].set_title('Cart Undersampled image')
axes[1, 1].imshow(cart_us_img, cmap='gray')

axes[0, 2].set_title('Radial Undersampled kspace')
axes[0, 2].imshow(radial_us_kspace, cmap='gray')

axes[1, 2].set_title('Radial Undersampled image')
axes[1, 2].imshow(radial_us_img, cmap='gray')

axes[0, 3].set_title('Random Undersampled kspace')
axes[0, 3].imshow(random_us_kspace, cmap='gray')

axes[1, 3].set_title('Random Undersampled image')
axes[1, 3].imshow(random_us_img, cmap='gray')
```

## Total variation reconstruction

```python
def tv_loss(img):
     # compute image gradient in two directions
    tv_h = torch.abs(img[1:, :] - img[:-1, :]).mean()
    tv_w = torch.abs(img[:, 1:] - img[:, :-1]).mean() 
    return tv_h + tv_w

# img -> kspace -> mask -> undersampled_img
radial_us_img = torch.abs(ifft(radial_mask * fft(img)))
radial_us_img = normalize(radial_us_img, radial_us_img.min(), radial_us_img.max() - radial_us_img.min())

tv_coef = 0.5  # 0.5
lr = 1e-3
iterations = 1000

recon_img = torch.zeros_like(random_us_img)
recon_img.requires_grad_()
losses = []

# optimization process
for i in range(iterations):
    loss = (recon_img - radial_us_img).pow(2).mean() + tv_coef * tv_loss(recon_img)
    loss.backward()
    loss = loss.item()
    losses.append(loss)

    with torch.no_grad():
        recon_img.sub_(lr * recon_img.grad)
        recon_img.grad.zero_()

        
# present the result and loss curve
fig, axes = plt.subplots(1, 2, figsize=(8, 4), layout='compressed')
axes[0].set_title('Reconstructed image')
axes[0].imshow(recon_img.detach().numpy(), cmap='gray')
axes[1].set_title('Loss curve')
axes[1].plot(list(range(len(losses))), losses)
```

## Simple SENSE

```python
def Simple_SENSE(sw_cu_img, csm, R, show=True):
    # C is the number of coils, R is the acceleration rate
    C, H, W = sw_cu_img.shape
    recon_img = torch.zeros(H, W).type(torch.complex64)
    for col in range(0, W//R):
        # select R positions in a row
        col_idxes = list(range(col, W, W//R))
        for row in range(0, H):
            # solve a sub-problem to get R pixels
            recon_img[row, col_idxes] = pinv(csm[:, row, col_idxes]) @ sw_cu_img[:, row, col]
        
        # visualize the reconstruction process
        if show and (col % 3 == 0 or col == H//R - 1): 
            plt.imshow(torch.abs(recon_img), cmap='gray')
            display.display(plt.gcf())
            display.clear_output(wait=True)
    
    return recon_img
```

```python
acceleration = 2
n_coil = 4

# get cartesian mask according to the acceleration rate
cart_mask = torch.zeros_like(img)
cart_mask[:, 0:-1:acceleration] = 1

# weighted_img -> kspace -> mask -> undersampled_img
sw_cu_img = ifft(cart_mask * fft(csm * img))

# present undersampled_img
fig, axes = plt.subplots(3, 5, layout='compressed', figsize=(12, 8))
for i, row in enumerate(axes):
    for j, col in enumerate(axes[i]):
        axes[i, j].imshow(torch.abs(sw_cu_img)[3*i + j], cmap='gray')
```

```python
# reconstruct
recon_img = Simple_SENSE(sw_cu_img[:n_coil, :, :], csm[:n_coil, :, :], acceleration)

plt.clf()

# compare results and original image
fig, axes = plt.subplots(1, 2, layout='compressed', figsize=(8, 4))
axes[0].set_title('Reconstructed image')
axes[0].imshow(torch.abs(recon_img), cmap='gray')
axes[1].set_title('Original image')
axes[1].imshow(img, cmap='gray')
```
