```python
import numpy as np
import matplotlib.pyplot as plt
from phantominator import shepp_logan
from skimage.util import view_as_windows
```

```python
N = 128
ncoils = 4
xx = np.linspace(0, 1, N)
x, y = np.meshgrid(xx, xx)
mps = np.zeros((N, N, ncoils))
mps[..., 0] = x**2
mps[..., 1] = 1 - x**2
mps[..., 2] = y**2
mps[..., 3] = 1 - y**2

fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(12, 4))
axes[0].imshow(mps[..., 0], cmap='gray')
axes[1].imshow(mps[..., 1], cmap='gray')
axes[2].imshow(mps[..., 2], cmap='gray')
axes[3].imshow(mps[..., 3], cmap='gray')
plt.show()
```

```python
ph = shepp_logan(N)
plt.imshow(ph, cmap='gray')
```

```python
imspace = ph[..., None] * mps
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(12, 4))
axes[0].imshow(imspace[..., 0], cmap='gray')
axes[1].imshow(imspace[..., 1], cmap='gray')
axes[2].imshow(imspace[..., 2], cmap='gray')
axes[3].imshow(imspace[..., 3], cmap='gray')
plt.show()
```

```python
imspace = imspace.astype('complex')
ax = (0, 1)
kspace = np.fft.fftshift(np.fft.fft2(np.fft.ifftshift(imspace, axes=ax), axes=ax), axes=ax)
plt.imshow(abs(kspace[..., 0]), cmap="gray", vmax=20)
```

```python
mask = np.zeros((N, N, ncoils), dtype=bool)
mask[:, ::2, :] = 1
ac_pad = 10
mask[:, N//2 - ac_pad: N//2 + ac_pad, :] = 1
plt.imshow(mask[..., 0], cmap="gray")
```

```python
kspace_masked = kspace * mask
plt.imshow(abs(kspace_masked[..., 0]), cmap="gray", vmax=20)
```

```python
kspace_masked_ifft = np.fft.fftshift(np.fft.ifft2(np.fft.ifftshift(kspace_masked, axes=ax), axes=ax), axes=ax)
plt.imshow(abs(kspace_masked_ifft[..., 0]), cmap="gray")
```

```python
ac_region = kspace_masked[:, N//2 - ac_pad: N//2 + ac_pad, :]
plt.imshow(abs(ac_region[..., 0]), cmap="gray", vmax=20)
```

```python
kernel_size = [5, 5]
kx, ky = kernel_size
kspace_masked = np.pad(kspace_masked, ((kx//2, kx//2), (ky//2, ky//2), (0, 0)), mode='constant')
ac_region = np.pad(ac_region, ((kx//2, kx//2), (ky//2, ky//2), (0, 0)), mode='constant')
mask = np.pad(mask, ((kx//2, kx//2), (ky//2, ky//2), (0, 0)), mode='constant')
```

```python
mask_patches = view_as_windows(mask[..., 0], kernel_size)
mask_patches = mask_patches.reshape((-1, kx, ky))
# Find the unique patches and associate them with indices
mask_patches, iidx = np.unique(mask_patches, return_inverse=True, axis=0)
# Filter out geometries that don't have a hole at the center.
# These are all the kernel geometries we actually need to compute weights for.
valid_index = np.argwhere(~mask_patches[:, kx//2, ky//2]).squeeze()
# visualize valid mask patches
fig, axes = plt.subplots(4, 5, figsize=(12, 8))
for i, axe in enumerate(axes.flatten()):
    axe.imshow(mask_patches[valid_index[i]], cmap="gray")
```
